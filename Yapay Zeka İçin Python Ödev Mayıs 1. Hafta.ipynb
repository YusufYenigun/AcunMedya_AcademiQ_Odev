## Davies-Bouldin (DB) İndeksi, kümeleme algoritmalarının performansını değerlendirmek için kullanılan bir iç geçerlik metriğidir. Küme sayısı belirlenirken veya algoritmalar karşılaştırılırken oldukça yararlıdır.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
import matplotlib.pyplot as plt

# Örnek veri üret
X, y = make_blobs(n_samples=300, centers=4, cluster_std=0.6, random_state=42)

# KMeans ile kümeleme
kmeans = KMeans(n_clusters=4, random_state=42)
labels = kmeans.fit_predict(X)

# Davies-Bouldin skorunu hesapla
db_index = davies_bouldin_score(X, labels)
print("Davies-Bouldin Skoru:", db_index)

# Görselleştirme
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X')
plt.title("KMeans Kümeleme")
plt.show()
```

### Küme sayısını (k) belirlemek için. Birden fazla kümeleme algoritması veya parametresi karşılaştırılacaksa. Etiketli veri olmadan model performansı ölçülmek istendiğinde.
